```python
import numpy as np
import matplotlib.pyplot as plt

# 定义图片的大小
height, width = 160, 50

# 创建一个浅蓝色的RGB颜色数组
# '#ADD8E6' 对应的 RGB 值为 (173, 216, 230)
light_blue = np.array([173, 216, 230]) / 255.0  # 归一化到 0-1 之间

# 创建一个浅蓝色图片
image = np.ones((height, width, 3)) * light_blue

# 显示图片
plt.imshow(image)
plt.axis('off')  # 隐藏坐标轴
plt.savefig('light_blue_image.png', bbox_inches='tight', pad_inches=0, dpi=300)

plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# 读取图像
image_path = 'cat.jpg'
image = Image.open(image_path).convert('RGB')
image_np = np.array(image)/255.0

# 设置参数
num_time_steps = 100
noise_std_devs = np.linspace(0.01, 5.0, num_time_steps)  # 随时间步变化的噪声强度

# 创建图像噪声序列
noisy_images = []
for std_dev in noise_std_devs:
    noise = np.random.normal(loc=0, scale=std_dev, size=image_np.shape)
    noisy_image_np = np.clip(image_np + noise, 0, 1)
    noisy_images.append(noisy_image_np)

# 可视化部分噪声图像
fig, axes = plt.subplots(1, 5, figsize=(15, 15))
for ax, noisy_image_np in zip(axes.flatten(), noisy_images[::20]):  # 每隔10步显示一张图
    ax.imshow(noisy_image_np)
    ax.axis('off')
plt.savefig('noise_cat.png', bbox_inches='tight', pad_inches=0, dpi=300)

plt.show()
```
